```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from fredapi import Fred
```

# HOUSING PRICES

```python
file_name_yearly = 'yearly_housing_prices.csv'
housing_df = pd.read_csv(file_name_yearly)
housing_df = housing_df.set_index('year')
#print(housing_df.columns)
housing_df.drop('Unnamed: 0', axis=1, inplace=True)
housing_df.head()
```

# MINIMUM WAGE

```python
file = 'minimum_wage_data_clean.csv'
min_wage_df = pd.read_csv(file)
min_wage_df['year'] = pd.to_datetime(min_wage_df['year'])
```

```python
min_wage_df['year'] =  pd.to_datetime(min_wage_df['year'], format='%Y-%m-%d')
min_wage_df.tail()
```

# Past dollars in terms of recent dollars = Dollar amount Ã— Ending-period CPI Ã· Beginning-period CPI.

```python
fred = Fred(api_key='83dd44fe528f2b9ef86fd007dc7235f7')
cpi = fred.get_series('CPALTT01USA661S')
# start at 1968
cpi = cpi[13:]
cpi_df = pd.DataFrame({'year': cpi.index, 'cpi': cpi.values})

#cpi_df.dtypes

cpi_df.head()
```

# CPI 
In order for the past prices of homes and wages to be understood as a current market value, I will use the CPI to convert them to 2018 prices.

```python
cpi_2018 = 105.944651
# merge cpi table with min_wage_df
combine_df = pd.merge(min_wage_df, cpi_df, how='inner', on='year')
combine_df.head()
```

```python
combine_df = combine_df.set_index('year')
combine_df.head()
```

```python
combine_df_temp = combine_df
#combine_df_temp.head()
combine_df_temp.head()
```

```python
cols = combine_df_temp.columns

for c in cols[:-1]:
    combine_df_temp[c] = combine_df_temp[c] * cpi_2018 / combine_df_temp['cpi']

combine_df_temp.head()
```

# MINIMUM WAGE ADJUSTED 
This graph shows the Federal Minimum Wage adjusted for inflation. You can clearly see the downward trend. If I had not adjusted the minimum wage for inflation, the graph would have instead steadly rose, which is not represenative of what the real trend is. 

```python
combine_df_temp.reset_index().plot(x='year', y='FED')
plt.xlabel('year')
plt.ylabel('FED')
plt.show()
```

# ADJUST HOUSING WITH CPI

```python
housing_df2 = housing_df.reset_index()
housing_df2['year'] = pd.to_datetime(housing_df2['year'], format='%Y-%m-%d')
housing_df2.head()
```

```python
cpi2 = fred.get_series('CPALTT01USA661S')
#start at 1975
cpi2 = cpi2[20:]
cpi2_df = pd.DataFrame({'year': cpi2.index, 'cpi': cpi2.values})
cpi2_df['year'] = pd.to_datetime(cpi2_df['year'], format='%Y-%m-%d')
```

```python
cpi_2018 = 105.944651

# merge cpi table with min_wage_df
c_df = pd.merge(housing_df2, cpi2_df, how='inner', on='year')
c_df.head()
```

```python
c_df = c_df.set_index('year')
c_df.head()
```

```python
temp_c_df = c_df 
```

```python
cols2 = temp_c_df.columns

for c in cols2[:-1]:
    temp_c_df[c] = temp_c_df[c] * cpi_2018 / temp_c_df['cpi']

temp_c_df.rename(columns={'United States not seasonally adjusted':'FED'}, inplace=True)
temp_c_df.head()
```

# HOUSING PRICES ADJUSTED
This is the average cost of a one family home in the United States adjusted for inflation. The housing bubble is pretty clearly defined in the 2000-2010 range of the graph, and you can see that starting in 2012, it is rising again. It is also valuable to see that from 1975-2000, the average house cost around 130,000 in 2018 money. It is only in the last two decades housing costs rose substatially. 

```python
temp_c_df.reset_index().plot(x='year', y='FED')
plt.xlabel('year')
plt.ylabel('FED Average ($)')
plt.show()
```

# COMPARING HOUSING PRICES AND MINIMUM WAGE

Plotting the adjusted housing prices and minimum wages together on the same plot show how these two figures interact with each other. Most of the time, they seem to be moving in opposite directions. When housing costs go up, minimum wage trends down and vice versa. 

```python
# make new dataframe
# cut minimum wage to '75

housing_vs_wage_df = pd.DataFrame([combine_df_temp.index[7:], temp_c_df.FED, combine_df_temp.FED[7:]])
housing_vs_wage_df = housing_vs_wage_df.T
housing_vs_wage_df.columns = ['year', 'housing cost', 'minimum wage']
housing_vs_wage_df = housing_vs_wage_df.set_index('year')
housing_vs_wage_df.head()
```

```python
figure = plt.figure()
ax1 = figure.add_subplot(111)
# housing cost
ax1.plot(housing_vs_wage_df['housing cost'], color='b', label='housing')
ax2 = ax1.twinx()
# mininum wage
ax2.plot(housing_vs_wage_df['minimum wage'], color='r', label='minimum wage')
ax2.grid(None)
ax1.set(xlabel='years', ylabel='100,000 ($)')
ax2.set(ylabel='($)')
plt.legend( loc='upper left')
plt.title('Minimum Wage VS Median House Value Adjusted by CPI')
plt.show()
```
