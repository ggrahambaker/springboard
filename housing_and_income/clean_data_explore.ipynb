MINIMUM WAGE

```python
file = 'minimum_wage_data.csv'
min_wage_df = pd.read_csv(file)
cols = min_wage_df.columns.values
cols[0] = 'year'
min_wage_df.columns = cols
min_wage_df['year'] = pd.to_datetime(min_wage_df['year'])
min_wage_df.set_index('year', inplace=True)
min_wage_df.head()
```

```python
has_null = min_wage_df.columns[min_wage_df.isna().any()].tolist()

for col in has_null:
    #print(col)
    min_wage_df[col].fillna(min_wage_df['FED'], inplace=True)
#min_wage_df.fillna(min_wage_df.FED)
#min_wage_df[has_null].fillna(min_wage_df['FED'], inplace=True)
#min_wage_df[has_null].fillna(min_wage_df['FED'], inplace=True)
min_wage_df.head()
```
