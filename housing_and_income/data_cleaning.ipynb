housing!

```python
housing_prices_df = pd.read_csv('housing_prices.csv')
```

```python
housing_prices_df.head()
# 2018 - 1975 = 43
# split into 43!
```

```python
# need to take average of each year!
series_arr_of_years = []
past = 0
curr = 12
for chunk in range(0, 44):
    series_arr_of_years.append(housing_prices_df[past:curr].mean())
    past += 12
    curr += 12

yearly_housing_price = pd.concat(series_arr_of_years, axis=1)
yearly_housing_price = yearly_housing_price.T
yearly_housing_price


# make year cols, start in 1975
col_names = []
for num in range(0, 44):
    col_names.append(1975 + num)

yearly_housing_price.index = col_names
yearly_housing_price.head()
```

```python
# save this as csv
file_name_yearly = 'yearly_housing_prices.csv'
yearly_housing_price.to_csv(file_name_yearly)
```
